i want to do visualising to see the average salary for each experience level

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate the average salary for each experience level
average_salary_by_experience = df.groupby('experience_level')['salary_in_usd'].mean().sort_values(ascending=False)

# Create the bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x=average_salary_by_experience.index, y=average_salary_by_experience.values, palette='viridis')
plt.title('Average Salary by Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Average Salary (USD)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

as expected, the more experience employee have higher average salary
